```
import logging
import time
import datetime
'''
Program used for custom logging in Pyspark and it will be creating a log file
'''
# Incase If You want to inlcude a timestamp with which you would create file follow up with the below statement
# file_date = datetime.datetime.fromtimestamp(time.time()).strftime('%Y-%m-%d-%H-%M-%S')
# p_filename = "custom_log"+file_date+".log"
p_dir = '/tmp/'
p_filename = 'custom_log.log'
p_logfile = p_dir + p_filename
print(p_logfile)
# Create logger with 'custom_log'
logger = logging.getLogger('custom_log')
logger.setLevel(logging.INFO)
# Create file handler with a log level
f_handler = logging.FileHandler(p_logfile, mode = 'a')
# Create console handler with a log level
c_handler = logging.StreamHandler()
c_handler.setLevel(logging.INFO)
# Create Formatter and add it to the handlers
f_format = logging.Formatter("%(asctime)s - %(name)s - %(levelname)s - %(message)s")
# Setting for ignoring frequent log information
# logging.getlogger("py4j").setLevel(logging.ERROR)
f_handler.setFormatter(f_format)
c_handler.setFormatter(f_format)
#Clearing old frequent log information to ignore that
if (logger.hasHandlers()):
  logger.handlers.clear()
# Add Handlers to the Logger
logger.addHandler(f_handler)
logger.addHandler(c_handler)
```
